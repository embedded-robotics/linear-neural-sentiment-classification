#### This program will plot the graphs for Exploration 1 in which we have tested Logistic Regression for different learning rates and different epochs

```python
import matplotlib.pyplot as plt
```

```python
epochs = [5, 10, 20, 30, 40]

# Accuracy for Learning Rate 0.1
train_lr_0_1 = [0.944, 0.975, 0.989, 0.994, 0.997]
dev_lr_0_1 = [0.791, 0.791, 0.787, 0.787, 0.786]

# Accuracy for Learning Rate 0.01
train_lr_0_01 = [0.836, 0.880, 0.918, 0.939, 0.951]
dev_lr_0_01 = [0.775, 0.778, 0.782, 0.787, 0.787]

# Accuracy for Learning Rate 0.001
train_lr_0_001 = [0.693, 0.728, 0.770, 0.799, 0.821]
dev_lr_0_001 = [0.693, 0.714, 0.745, 0.751, 0.763]
```

```python
plt.figure(figsize=(10,6))
plt.plot(epochs, train_lr_0_1, label='lr=0.1, train acc')
plt.plot(epochs, dev_lr_0_1, label='lr=0.1, dev acc')
plt.plot(epochs, train_lr_0_01, label='lr=0.01, train acc')
plt.plot(epochs, dev_lr_0_01, label='lr=0.01, dev acc')
plt.plot(epochs, train_lr_0_001, label='lr=0.001, train acc')
plt.plot(epochs, dev_lr_0_001, label='lr=0.001, dev acc')
plt.xlabel('No. of Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
